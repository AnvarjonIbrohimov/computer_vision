```python
import cv2
```

```python
from google.colab.patches import cv2_imshow  # shu bolmasa colab rasmni oqiy olmaydi

image = cv2.imread("/content/cat_image.jpeg") # faylni oqib berish uchun hizmat qiladi
cv2_imshow(image) # bu esa faylni oqish uchun hizmat qiladi
```

```python
print("SHAPE:", image.shape)
# 350 => height
# 625 => width
# 3   => color channels(kanallar) RGB => Red, Green, Blue <=> opencv da BGR deyiladi
print("Height:", image.shape[0])
print("Widtht:", image.shape[1])
print("Color:", image.shape[2])
```

```python
( b, g, r ) = image[ 100, 50]   # bu biz hohlagan pixelni qaysi RGBdagi ranglarini chiqarishimiz uchun
print("Blue:",b, "Green:",g, "Red:",r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray,(300,300))  # bu rasmni bniz hohlagan pexelga otkazish uchun
cv2_imshow(resized)
```

```python
cropped = image[80:280, 200:400]  # 80:280 => height boyicha  200:400 bu esa widht boyicha
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_180)  # bu rasmni 180 gradusga aylantirib beradi
cv2_imshow(rotated)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE) # 시계 방향
cv2_imshow(rotated)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_COUNTERCLOCKWISE) # bu esa 반시계방향
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)  # bu huddi oynada korgandek korsatadi yani rasmni teskari qilib beradi
cv2_imshow(flipped)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)   # bu shuncha gray holatga otkazilgan asl korinishi
cv2_imshow(gray)
```

```python
cv2.imwrite("flipped_cat.jpg", flipped)  # BU RASMNI SAHRANIT QILIB OLISH UCHUN
```

```python
cv2.imwrite("cropped_cat.jpg", cropped) # bu esa cropped(kesib olingan) holatdagisini sahranit qilib olishimiz mumkin
```
