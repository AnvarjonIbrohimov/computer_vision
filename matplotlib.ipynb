```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 30])

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

**# Multiple Lines in One Graph with Numpy and Matplotlib**

```python
from matplotlib.lines import lineStyles
x = np.array([1,2,3,4])
y = np.array([3, 6, 1, 3])

plt.plot(y, linestyle="dotted", color="aqua", marker="o")
plt.title("Styled Line Example")
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([5,2,9,6])
y2 = np.array([2, 8, 10, 5])

plt.plot(x, y1, label="Line 1", color="red", marker="s")
plt.plot(x, y2, label="Line 2", color="aqua", linestyle="--", marker='o')
plt.title("Multiple Lines Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()  # bu ekranda Line 1 va Line 2 qanday rangdaligi va qanday belgidaligi korinadi
plt.show()
```

**Bar Chart with Numpy and Matplotlib**

```python
x = np.array(["Akmal", "Bobur", "Nizom", "Dilmurod", "Akram", "Abror", "Misha", "Ali"])
y = np.array([88, 84, 65, 61, 92, 75, 53, 99])

plt.bar(x, y, color='aqua')
plt.title("Bar Chart Example")
plt.xlabel("Ismlar")
plt.ylabel("Ballar")
plt.show()
```

**Scatter Plot with Numpy and Matplotlib**

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 2, 4, 3, 6])

plt.scatter(x, y, color="purple", s=50)
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

# Histogram with Numpy and Matplotlib

```python
data = np.array([3, 4, 5, 2, 7, 8 ,9 ,10,12,1,3])

plt.hist(data, bins=6, color="skyblue", edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```

# Subplots(Multiple Graphs Together)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title("Plot 2")
plt.tight_layout()
plt.show()
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 8, 6, 10])

plt.plot(x, y, color="green")
plt.title("Save Figure Example")
plt.savefig('my_plot.png') # Saves in current folder
plt.show
```

```python

import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import axes3d

plt.style.use('_mpl-gallery')

# Make data
X, Y, Z = axes3d.get_test_data(0.05)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
np.random.seed(19680801)
n = 100
rng = np.random.default_rng()
xs = rng.uniform(23, 32, n)
ys = rng.uniform(0, 100, n)
zs = rng.uniform(-50, -25, n)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(xs, ys, zs)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap="Blues")

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
